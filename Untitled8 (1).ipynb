```python
import pandas as pd
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Function to handle negative values and zeros by replacing them with NaN
def handle_invalid_values(df, columns):
    for col in columns:
        # Replace values <= 0 with NaN
        df[col] = df[col].apply(lambda x: np.nan if x <= 0 else x)
    return df

# Function to compute peak electron density (Nm)
def compute_nm(F107, SN, Kp, C0=1e12, C1=1e10, C2=1e10, C3=-1e9):
    return C0 + C1 * F107 + C2 * SN + C3 * Kp

# Chapman function to model electron density
def chapman_function(h, Hm, Nm, H):
    """
    Compute the electron density at height h based on the Chapman function.

    Parameters:
    - h: Height (km)
    - Hm: Height of maximum ionization (km)
    - Nm: Peak electron density (m^-3)
    - H: Scale height (km)

    Returns:
    - Electron density at height h (m^-3)
    """
    z = (h - Hm) / H  # Scale height adjustment
    # Exponentially decay based on Chapman function
    return Nm * np.exp(1 - z - np.exp(-z))

# Function to compute TEC by integrating the Chapman function
def compute_tec(Nm, Hm=350, H=50, h1=90, h2=1000):
    """
    Compute Total Electron Content (TEC) by integrating the Chapman function over height.

    Parameters:
    - Nm: Peak electron density (m^-3)
    - Hm: Height of maximum ionization (km)
    - H: Scale height (km)
    - h1: Lower height limit for integration (km)
    - h2: Upper height limit for integration (km)

    Returns:
    - TEC in units of electrons per square meter (m^-2)
    """
    tec, _ = quad(chapman_function, h1, h2, args=(Hm, Nm, H))

    # Multiply by a scaling factor to increase TEC values (if necessary)
    scaling_factor = 1e3 # Adjust this factor to control the TEC range
    return tec * scaling_factor  # Return scaled TEC

# Load dataset (replace with your actual data file path)
file_path = "updated_tec_data.csv"  # Replace with your dataset path
df = pd.read_csv(file_path)

# Handle invalid values (negative and zero) in 'F10.7obs' and 'SN'
df = handle_invalid_values(df, ['F10.7obs', 'SN'])

# Step 2: Interpolate (linear interpolation) to estimate the missing values
df['F10.7obs'] = df['F10.7obs'].interpolate(method='linear', limit_direction='both')
df['SN'] = df['SN'].interpolate(method='linear', limit_direction='both')

# Step 3: Handle Missing or Negative Values in Kp (Kp1 to Kp8)
df['Kp_avg'] = df[['Kp1', 'Kp2', 'Kp3', 'Kp4', 'Kp5', 'Kp6', 'Kp7', 'Kp8']].mean(axis=1)

# Step 4: Compute TEC for Each Row in the Dataset
tec_values = []
for _, row in df.iterrows():
    F107 = row['F10.7obs']
    SN = row['SN']
    Kp_avg = row['Kp_avg']  # Use Kp_avg instead of Kp1

    # Compute Nm (Peak electron density)
    Nm = compute_nm(F107, SN, Kp_avg)

    # Compute TEC using the Chapman function
    tec = compute_tec(Nm)
    tec_values.append(tec)

# Add TEC values to the dataset
df['TEC (TECU)'] = tec_values

data['Kp_avg'] = data[['Kp1', 'Kp2', 'Kp3', 'Kp4', 'Kp5', 'Kp6', 'Kp7', 'Kp8']].mean(axis=1)


# Step 5: Monthly averages for visualization
# Ensure columns are integers
# Convert to datetime using the correct column names
# Combine the year, month, and day into a single column for datetime conversion
df['Date'] = pd.to_datetime(df[['YYY', 'MM', 'DD']].astype(str).agg('-'.join, axis=1))

# Group by month and calculate monthly averages
df['Month'] = df['Date'].dt.to_period('M')
monthly_avg = df.groupby('Month')[['F10.7obs', 'SN', 'TEC (TECU)']].mean()

print(monthly_avg)

# Step 6: Plot trends
# Plot Solar Activity and TEC in separate graphs

# F10.7 and Sunspot Number (SN)
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg.index.to_timestamp(), monthly_avg['F10.7obs'], label='F10.7 (Solar Flux)', color='blue')
plt.xlabel('Time')
plt.ylabel('F10.7obs (Solar Flux)')
plt.title('Monthly Solar Flux (F10.7)')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(monthly_avg.index.to_timestamp(), monthly_avg['SN'], label='Sunspot Number (SN)', color='orange')
plt.xlabel('Time')
plt.ylabel('SN (Sunspot Number)')
plt.title('Monthly Sunspot Number (SN)')
plt.legend()
plt.grid()
plt.show()

# TEC Values
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg.index.to_timestamp(), monthly_avg['TEC (TECU)'], label='TEC (TECU)', color='green')
plt.xlabel('Time')
plt.ylabel('TEC (TECU)')
plt.title('Monthly TEC (TECU)')
plt.legend()
plt.grid()
plt.show()

# Save the updated dataset with TEC values
df.to_csv("data.csv", index=False)
print("TEC calculation and preprocessing completed. Results saved to 'data_tec.csv'.")
```

```python
import numpy as np

# Function to prepare data for LSTM model
def prepare_data_for_lstm(df, feature_col, n_steps):
    X, y = [], []
    for i in range(n_steps, len(df)):
        X.append(df[feature_col].iloc[i-n_steps:i].values)
        y.append(df[feature_col].iloc[i])
    return np.array(X), np.array(y)

# Example: Prepare data for LSTM model
n_steps = 5  # Use last 5 days of data to predict the next value
X_train_SN, y_train_SN = prepare_data_for_lstm(df, 'SN', n_steps)
X_train_F107, y_train_F107 = prepare_data_for_lstm(df, 'F10.7obs', n_steps)
X_train_Kp, y_train_Kp = prepare_data_for_lstm(df, 'Kp_avg', n_steps)  # You can use any Kp column
X_train_TEC, y_train_TEC = prepare_data_for_lstm(df, 'TEC (TECU)', n_steps)

# Reshaping data to be compatible with LSTM input format (samples, timesteps, features)
X_train_SN = X_train_SN.reshape((X_train_SN.shape[0], X_train_SN.shape[1], 1))
X_train_F107 = X_train_F107.reshape((X_train_F107.shape[0], X_train_F107.shape[1], 1))
X_train_Kp = X_train_Kp.reshape((X_train_Kp.shape[0], X_train_Kp.shape[1], 1))
X_train_TEC = X_train_TEC.reshape((X_train_TEC.shape[0], X_train_TEC.shape[1], 1))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Function to create LSTM model
def create_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(50, return_sequences=False))
    model.add(Dense(1))  # Output layer (single value prediction)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Create LSTM models for each parameter (SN, F10.7, Kp, TEC)
model_lstm_SN = create_lstm_model((X_train_SN.shape[1], 1))
model_lstm_F107 = create_lstm_model((X_train_F107.shape[1], 1))
model_lstm_Kp = create_lstm_model((X_train_Kp.shape[1], 1))
model_lstm_TEC = create_lstm_model((X_train_TEC.shape[1], 1))

# Train models
model_lstm_SN.fit(X_train_SN, y_train_SN, epochs=20, batch_size=32)
model_lstm_F107.fit(X_train_F107, y_train_F107, epochs=20, batch_size=32)
model_lstm_Kp.fit(X_train_Kp, y_train_Kp, epochs=20, batch_size=32)
model_lstm_TEC.fit(X_train_TEC, y_train_TEC, epochs=20, batch_size=32)
```

```python
# Assume 'df' is your DataFrame with TEC and solar/geomagnetic parameters
# Example: You need to predict TEC using previous values of TEC and other features like F10.7, SN, Kp, etc.

# Drop any missing values and select relevant columns
df = df.dropna(subset=['TEC (TECU)', 'F10.7obs', 'SN', 'Kp1', 'Kp2', 'Kp3', 'Kp4', 'Kp5', 'Kp6', 'Kp7', 'Kp8'])

# Select features (X) and target variable (y)
X = df[['F10.7obs', 'SN', 'Kp1', 'Kp2', 'Kp3', 'Kp4', 'Kp5', 'Kp6', 'Kp7', 'Kp8']]  # Example features
y = df['TEC (TECU)']  # Target variable

# Scale the data (LSTM models work better with scaled data)
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1))

# Reshape X to be 3D: [samples, time steps, features]
# For time-series, this means creating sequences of past values to predict the next value
def create_sequences(data, sequence_length=10):
    X_seq = []
    y_seq = []

    for i in range(len(data) - sequence_length):
        X_seq.append(data[i:i+sequence_length])
        y_seq.append(data[i+sequence_length])

    return np.array(X_seq), np.array(y_seq)

# Create sequences
sequence_length = 10
X_seq, y_seq = create_sequences(X_scaled, sequence_length)
```

```python
# Split into train and test sets (e.g., training on 80% and testing on 20%)
X_train, X_test, y_train, y_test = train_test_split(X_seq, y_seq, test_size=0.2, shuffle=False)
```

```python
# Build the LSTM model
model = Sequential()

# First LSTM layer with dropout for regularization
model.add(LSTM(units=50, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting

# Second LSTM layer with dropout
model.add(LSTM(units=50, activation='relu', return_sequences=False))
model.add(Dropout(0.2))

# Output layer
model.add(Dense(units=1))  # Single output for predicting TEC

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Summarize the model
model.summary()
```

```python
# Train the LSTM model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Make predictions
predicted_y = model.predict(X_test)

# Inverse transform the predictions and actual values
predicted_y_actual = scaler_y.inverse_transform(predicted_y)
y_test_actual = scaler_y.inverse_transform(y_test)

# Print the first few predictions and actual values for comparison
print("Predicted TEC values (inverse scaled): ", predicted_y_actual[:5])
print("Actual TEC values (inverse scaled): ", y_test_actual[:5])
```
